```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('file.csv')
df
```

```python

df['no_of_crimes'] = df['no_of_crimes'].astype(np.int64)
df.dtypes
```

```python

#Q. 1) Convert the Datatype of 'Date' column to Date-Time format.
df['date'] = pd.to_datetime(df['date'],errors='coerce')

df
```

```python
#Q. 2) Add a new column ''year'' in the dataframe, which contains years only.
df['year'] = df['date'].dt.year

df
```

```python
#.B) Add a new column ''month'' as 2nd column in the dataframe, which contains month only.

df['month'] = df['date'].dt.month_name()

df = df[['year','month','date','area','average_price','code','houses_sold','no_of_crimes']]
df
```

```python
#Q. 3) Remove the columns 'year' and 'month' from the dataframe.
df =df.drop(columns=['year','month'])
df
```

```python
#Q. 4) Show all the records where 'No. of Crimes' is 0. And, how many such records are there ?

df = df.fillna(0)
grouped = df.groupby(['area'])['no_of_crimes'].mean().reset_index()
grouped['no_of_crimes'] = grouped['no_of_crimes'].astype(np.int64)
grouped[grouped['no_of_crimes']==0]
```

```python
#Q. 5) What is the maximum & minimum 'average_price' per year in england ?
max_average_price = df['average_price'].max()
min_average_price = df['average_price'].min()

max_index = df['average_price'].idxmax()
min_index = df['average_price'].idxmin()

max_index = df.loc[max_index, 'area']
min_index = df.loc[min_index, 'area']

f'Maximum average price per year in england is {max_average_price} in {max_index} and the Minium average price per year in england is {min_average_price} in {min_index}' 

#area_ma
```

```python
#Q. 6) What is the Maximum & Minimum No. of Crimes recorded per area ?

max_crimes_per_area = df.groupby('area')['no_of_crimes'].max()
min_crimes_per_area = df.groupby('area')['no_of_crimes'].min()

result_df = pd.DataFrame({
    'area': max_crimes_per_area.index,
    'max_no_of_crimes': max_crimes_per_area.values,
    'min_no_of_crimes': min_crimes_per_area.values
})

result_df
```

```python
#Q. 7) Show the total count of records of each area, where average price is less than 100000.

result = df.groupby('area')['average_price'].mean().reset_index()
result['average_price'] = result['average_price'].astype(np.int64)
result[result['average_price']<100000]
```
