```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

sns.set(style='darkgrid')
df = sns.load_dataset('tips')

df
```

```python
#Grafica de barras de propinas promedio por sexo
color = ["lightblue",'lightgreen']

sns.barplot(data=df,x='sex',y='tip', errorbar=None, hue='sex')
plt.title("Propinas promedio por sexo")
plt.show()

#Hombres en promedio pagarian mas propina
```

```python
sns.barplot(data=df,x='day',y='tip',errorbar=None,hue='day')
plt.title("Propinas promedio por dia")
plt.show()

#Concluimos que el domingo se paga mas propina
```

```python
sns.barplot(data=df,x='time',y='tip',errorbar=None,hue='time')
plt.title("Propinas por tiempo")

#Concluimos que en las cenas se paga mas propina
```

```python
#Propinas promedio por fumador
sns.barplot(data=df , x='smoker',y='tip',errorbar=None,hue='smoker')

#Concluimos que si el cliente fuma o no, no afecta la propina
```

```python
#Grafico de barras de propinas promedio (dia,sexo)

sns.barplot(data=df,x='day',y='tip',errorbar=None,hue='sex')
```

```python
#Grafico de barras de propinas promedio (tiempo,sexo)

sns.barplot(data=df,x='time',y='tip',errorbar=None,hue='sex')
```

```python
#Grafico de barras de propinas promedio (fumador,sexo)

sns.barplot(data=df,x='smoker',y='tip',errorbar=None,hue='sex')
plt.title("Propinas promedio por Fumador/No Fumador y sexo")
plt.show()
```

```python
#Grafico de dispersion entre size y tip
sns.scatterplot(data=df,x='size',y='tip', hue='sex')
plt.title("Relacion entre tamaÃ±o de fiestra y propina dejada")
```

```python
#Grafico de barras de propinas promedio por tamaÃ±o de la fiesta
sns.barplot(data=df,x='size',y='tip', errorbar=None,hue='size')
plt.title('Propinas promedio por tamaÃ±o de la fiesta')
```

```python
#Calcular el porcentaje de propina con respecto al total y lo agregamos al dataframe

df['tip_percentage'] = (df['tip']/df['total_bill'])*100
df
```

```python
#Grafico de barras de porcentaje de propina promedio por sexo

sns.barplot(data=df,x='sex',y='tip_percentage', errorbar=None,hue='sex')
plt.title('Porcentaje de propina promedio por sexo')
```

```python
#Grafico de porcentaje de propina promedio por dia
sns.barplot(data=df,x='day',y='tip_percentage',errorbar=None,hue='day')
plt.title('Porcentaje de propina promedio por dia')
```

```python
#Grafico de porcentaje de propina promedio por dia
sns.barplot(data=df,x='time',y='tip_percentage',errorbar=None,hue='time')
plt.title('Porcentaje de propina promedio por tiempo')
```

```python
#Grafico de porcentaje de propina promedio por fumador/nofumador
sns.barplot(data=df,x='smoker',y='tip_percentage',errorbar=None,hue='smoker')
plt.title('Porcentaje de propina promedio por Fumador/NoFumador')
```
