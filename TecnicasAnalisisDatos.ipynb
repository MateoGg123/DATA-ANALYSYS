```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.decomposition import PCA
```

```python
#TECNICA PCA ES PARA PASAR DE UN GRUPO DE DIMENSIONES ( COLUMNAS )
#A UNO MENOR EJEMPLO DE 4 COLUMNAS A 2
df= sns.load_dataset('iris')
df.iloc[:,:-1]
```

```python
pca_model = PCA(n_components=2)
def_pca = pca_model.fit_transform(df.iloc[:,:-1])
def_pca
```

```python
#[:,0] = Todas las filas de la primera componente
#[:,1] = Todas las filas de la segunda componente

plt.scatter(def_pca[:,0],def_pca[:,1], c=df['species'].astype('category').cat.codes)
plt.xlabel("PCA1")
plt.ylabel("PCA2")
```

```python
#TECNICA TRANSFORMACION DE VARIABLES
sns.displot(data=df, x='sepal_length', hue='species', kde=True)
plt.show()
```

```python
#Transformacion Logaritmica 
#Ahora vemos los mismos datos pero con logaritmo aplicado a cada uno

df['sepal_length_log'] = np.log(df['sepal_length'])

sns.displot(data=df, x='sepal_length_log', hue='species', kde=True)
plt.show()
```

```python
#TECNICA NORMALIZACION DE DATOS
from sklearn.preprocessing import StandardScaler,MinMaxScaler
```

```python
#NOS DICE DE CUANTO ES LA DESVIACION ESTANDAR CON RESPECTO A
#LA MEDIA Y EL MINIMO Y MAXIMO

std_scaler = StandardScaler()
df_std_scaler = std_scaler.fit_transform(df.iloc[:,:-2])
df_std_scaler
```

```python
mm_scaler = MinMaxScaler()
df_normalized = mm_scaler.fit_transform(df.iloc[:,:-2])
df_normalized
```

```python
#TECNICA  DE COREELACION DE PEARSON
#PARA VER QUE TANTO SE CORRELACIONAN LAS VARIABLES

from scipy.stats import pearsonr
```

```python
#Calcular la correlacion

correlacion,_ = pearsonr(df['sepal_length'],df['petal_length'])
print(f'La correlacion de pearson entre el sepal y el petal length es {correlacion}')
```

```python
#TECNICA ANALISIS DE CLUSTERS
#Intenta separar elementos parecidos a los que no se parecen

from sklearn.cluster import KMeans

#N_clusters = numero de categorias que tenemos en nuestra tabla (3tipos de flores)
model = KMeans(n_clusters=3, random_state=42)

df["cluster"] = model.fit_predict(df.iloc[:,:-2])
df
```

```python
sns.scatterplot(data =df, x='sepal_length', y="petal_length",hue='cluster')
```
